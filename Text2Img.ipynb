```python
from diffusers import StableDiffusionPipeline
import torch

# Function to generate the image with error handling
def generate_image(prompt, model_name="stabilityai/stable-diffusion-2-1", height=768, width=768,
                   num_inference_steps=100, guidance_scale=9.0, seed=None):
    try:
        # Load the specified model
        print(f"Loading model: {model_name} ...")
        pipe = StableDiffusionPipeline.from_pretrained(model_name).to("cuda")

        # Set random seed for reproducibility if provided
        generator = None
        if seed is not None:
            generator = torch.Generator(device="cuda").manual_seed(seed)

        # Generate the image with the provided prompt and settings
        print(f"Generating image for prompt: '{prompt}' at {width}x{height} resolution...")
        image = pipe(prompt,
                     height=height,
                     width=width,
                     num_inference_steps=num_inference_steps,
                     guidance_scale=guidance_scale,
                     generator=generator).images[0]

        # Save and display the image
        image_file = "generated_image.png"
        image.save(image_file)
        image.show()
        print(f"Image successfully saved as {image_file}")

    except Exception as e:
        print(f"Error generating image: {e}")

# Main script to run the generation process
if __name__ == "__main__":
    # Define your prompt and parameters
    prompt = "A majestic landscape with mountains and a river under a starry sky"

    # Example 1: Using Stable Diffusion 2.1
    generate_image(prompt, model_name="stabilityai/stable-diffusion-2-1", height=1024, width=1024,
                   num_inference_steps=150, guidance_scale=9.0, seed=42)

    # Example 2: Using Stable Diffusion 1.4
    generate_image(prompt, model_name="CompVis/stable-diffusion-v1-4", height=768, width=768,
                   num_inference_steps=100, guidance_scale=8.5, seed=123)
```
